```python
import pandas as pd
import numpy as np
```

# Importing SEIR csv files from Matlab

```python
df1 = pd.read_csv('test1.csv',header=None)
df2 = pd.read_csv('test2.csv',header=None)
df3 = pd.read_csv('test3.csv',header=None)
df4 = pd.read_csv('test4.csv',header=None)
df5 = pd.read_csv('test5.csv',header=None)
dt1 = pd.read_csv('test1_0.9beta.csv',header=None)
dt2 = pd.read_csv('test2_0.9beta.csv',header=None)
dt3 = pd.read_csv('test3_0.9beta.csv',header=None)
dt4 = pd.read_csv('test4_0.9beta.csv',header=None)
dt5 = pd.read_csv('test5_0.9beta.csv',header=None)
df_1 = pd.read_csv('test1_0.8beta.csv',header=None)
df_2 = pd.read_csv('test2_0.8beta.csv',header=None)
df_3 = pd.read_csv('test3_0.8beta.csv',header=None)
df_4 = pd.read_csv('test4_0.8beta.csv',header=None)
df_5 = pd.read_csv('test5_0.8beta.csv',header=None)

fast3_1 = pd.read_csv('1_fast3.csv',header=None)
fast3_2 = pd.read_csv('2_fast3.csv',header=None)
fast3_3 = pd.read_csv('3_fast3.csv',header=None)
fast3_4 = pd.read_csv('4_fast3.csv',header=None)
fast3_5 = pd.read_csv('5_fast3.csv',header=None)

fast5_1 = pd.read_csv('1_fast5.csv',header=None)
fast5_2 = pd.read_csv('2_fast5.csv',header=None)
fast5_3 = pd.read_csv('3_fast5.csv',header=None)
fast5_4 = pd.read_csv('4_fast5.csv',header=None)
fast5_5 = pd.read_csv('5_fast5.csv',header=None)

delay5_1 = pd.read_csv('1_delay5.csv',header=None)
delay5_2 = pd.read_csv('2_delay5.csv',header=None)
delay5_3 = pd.read_csv('3_delay5.csv',header=None)
delay5_4 = pd.read_csv('4_delay5.csv',header=None)
delay5_5 = pd.read_csv('5_delay5.csv',header=None)


actual = pd.read_csv('Actual Data.csv', index_col = 0)
```

```python
df= pd.concat([df1,df2,df3,df4,df5],axis=1)
df.head()
dt = pd.concat([dt1,dt2,dt3,dt4,dt5],axis=1)
#delay = pd.concat([delay_1,delay_2,delay_3,delay_4,delay_5], axis=1)
df_0 = pd.concat([df_1,df_2,df_3,df_4,df_5],axis=1)
fast3 = pd.concat([fast3_1,fast3_2,fast3_3,fast3_4,fast3_5],axis=1)
fast5 = pd.concat([fast5_1,fast5_2,fast5_3,fast5_4,fast5_5],axis=1)
delay5 = pd.concat([delay5_1,delay5_2,delay5_3,delay5_4,delay5_5],axis=1)
```

```python
df.columns = ['Model T','Model S','Model Exposed','Model Infected','Model Removed']
dt.columns = ['Model T','Model S','Model Exposed w 10% Reduction in β','Model Infected w 10% Reduction in β','Model Removed']
df_0.columns = ['Model T','Model S','Model Exposed w 20% Reduction in β','Model Infected w 20% Reduction in β','Model Removed']
fast3.columns = ['Model T','Model S','Model Exposed w 3 day faster measures','Model Infected w 3 day faster measures','Model Removed']
fast5.columns = ['Model T','Model S','Model Exposed w 5 day faster measures','Model Infected w 5 day faster measures','Model Removed']
delay5.columns = ['Model T','Model S','Model Exposed w 5 day delay','Model Infected w 5 day delay','Model Removed']
```
